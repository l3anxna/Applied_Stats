# 8.1

```python
import numpy as np
import scipy.stats as sts
```

```python
lambda0 = 1
n = 20
alpha = 0.05
Nsim = 10000
```

```python
z_critical = sts.norm.ppf(1 - alpha/2)
```

```python
reject_count = 0
```

```python
for _ in range(Nsim):

    X = np.random.poisson(lam=lambda0, size=n)
    lambda_hat = np.mean(X)

    if lambda_hat == 0:
        continue

    W = (lambda_hat - lambda0) / np.sqrt(lambda_hat / n)
    if abs(W) > z_critical:
        reject_count += 1

estimated_alpha = reject_count / Nsim

print("Estimated Type I Error Rate =", estimated_alpha)
print("reject_count =", reject_count)
```

# 8.2

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
X = np.array([
93, 100, 102, 107, 98, 102, 107, 95, 93, 100,
109, 97, 104, 102, 107, 103, 99, 105, 102, 98,
100, 93, 100, 101, 101, 101, 97, 92, 104, 103,
106, 99, 99, 108, 107, 89, 100, 101, 95, 100,
100, 102, 109, 109, 109, 102, 100, 96, 116, 96
])

Y = np.array([
106, 100, 95, 101, 105, 108, 110, 103, 105, 113,
102, 107, 98, 109, 102, 100, 109, 113, 113, 93,
105, 98, 106, 106, 99, 109, 102, 108, 106, 103
])
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)

sd_x = np.std(X, ddof=1)
sd_y = np.std(Y, ddof=1)

sem_x = sts.sem(X)
sem_y = sts.sem(Y)

print('Sample X')
print('Mean =', mean_x)
print('SD =', sd_x)
print('SEM =', sem_x)

print('\nSample Y')
print('Mean =', mean_y)
print('SD =', sd_y)
print('SEM =', sem_y)
```

```python
plt.figure(figsize=(10,6))

plt.hist(X, bins=10, alpha=0.6, label='X sample')
plt.hist(Y, bins=10, alpha=0.6, label='Y sample')

plt.axvline(mean_x, linestyle='--', label='Mean X')
plt.axvline(mean_y, linestyle='--', label='Mean Y')

plt.xlabel('Thickness')
plt.ylabel('Frequency')
plt.title('Histogram of plank thickness')
plt.legend()
plt.show()
```

```python
result = sts.ttest_ind(X, Y, equal_var=True)

print('\nTwo-sample t-test')
print('Test statistic =', result.statistic)
print('p-value =', result.pvalue)
```
