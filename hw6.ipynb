```python
import numpy as np
import scipy.stats as sts
```

```python
alpha = 0.001
t_c = sts.norm.ppf(alpha)
print(f"Alpha: {alpha}, t_c: {t_c}")
```

```python
n = 300
alpha = 0.001
t_c = sts.norm.ppf(alpha)

p0 = 0.5
p1 = 0.4

se_p1 = np.sqrt(p1 * (1 - p1) / n)
mu = (p1 - p0) / se_p1
beta = sts.norm(mu, 1).cdf(t_c)
print(f"Beta: {beta}")
```
